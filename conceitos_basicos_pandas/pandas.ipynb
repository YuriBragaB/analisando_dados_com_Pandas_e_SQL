```python
import pandas as pd
import numpy as np
```

COLUNAS:

```python
labels = ['a', 'b', 'c']
minha_lista = [10, 20, 30]
arr = np.array([10, 20, 30])
d = {'a': 10, 'b':20, 'c': 30}
```

```python
pd.Series(labels)
# faz um serie
# sempre dá índices
```

```python
pd.Series(data = labels, index = minha_lista)
# é possível passar o que vai ser o índice
```

```python
pd.Series(labels, minha_lista)
```

```python
pd.Series(d)
# se já passar um dic deixa os índices = chaves, valores = valores
# pandas conversa muito bem com dicionários
```

```python
ser1 = pd.Series([1, 2, 3, 4], index = ['EUA', 'Alemanha', 'Rússia', 'Japão'])
```

```python
ser1
# index é uma espécie de seletor, é um ponto central para a manipulação de dados
```

```python
ser2 = pd.Series([1, 2, 5, 4], index = ['EUA', 'Alemanha', 'Italia', 'Japão'])
```

```python
ser2
```

```python
ser1[['EUA', 'Alemanha']]
# sae quiser pegar só um índice passe uma string
# se quiser pegar mais de um passe um array
```

```python
ser1 + ser2
# é possível fazer uma soma entre series, elas são orientadas pelo índice, se não aparecer nos dois arrays é retornado como valor NaN(not a number)
```

DATAFRAMES:

```python
from numpy.random import randn
# numpy é uma biblioteca com algebra
```

```python
df = pd.DataFrame(randn(5, 4), index = ['A', 'B', 'C', 'D', 'E'], columns='W X Y Z'.split( ))
# cada valor é a composição de um índece com uma coluna
# coluna = parte de cima
# índice = parte lateral
```

```python
df
# fica bem, mais organizado
```

```python
type(df)
```

```python
df['W']
# retorna um series
# diferente das series, é passado a coluna, não o índice
# dataframe = um conjunto de uma series
```

```python
df[['W']]
# devolver um dataframe
# indo contra o padrão de devolver em Series

```

```python
df['new'] = df['W']
# cria uma nova Serie com as informações da coluna W
```

```python
df
```

```python
df['new2'] = df['W'] + df['Y']
```

```python
df
```

```python
del df['new']
# deleta uma Serie(coluna)
```

```python
df
```

```python
df.drop('new2', axis= 1)
# axis = 1 = colunas
# axis = 0 = linhas
# retorna um novo dataframe, n modifica o original, caso n modifique o inplace, que é por padrão False
```

```python
df.drop('new2', axis= 1, inplace=True)
# inplace = True , altera o dataframe original, várias funções em pandas contém essa funcionalidade
```

```python
df
```

```python
df.loc['A']
# igual as outras funções da biblioteca pandas se passar o argumento em um array é possível ver mais de uma info
# se passar um array retorna um DataFrame
# loc, procura um A nas linhas
# devolve uma Series
```
