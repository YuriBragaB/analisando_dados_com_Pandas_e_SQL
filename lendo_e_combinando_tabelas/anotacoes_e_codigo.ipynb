### Merge, concat e join:

```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
df1 = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3'],
    'C' : ['C0', 'C1', 'C2', 'C3'],
    'D' : ['D0', 'D1', 'D2', 'D3']
    },
    index=[0, 1, 2, 3])
```

```python
df2 = pd.DataFrame({
    'A' : ['A4', 'A5', 'A6', 'A7'],
    'B' : ['B4', 'B5', 'B6', 'B7'],
    'C' : ['C4', 'C5', 'C6', 'C7'],
    'D' : ['D4', 'D5', 'D6', 'D7']
    },
    index=[4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({
    'A' : ['A8', 'A9', 'A10', 'A11'],
    'B' : ['B8', 'B9', 'B10', 'B11'],
    'C' : ['C8', 'C9', 'C10', 'C11'],
    'D' : ['D8', 'D9', 'D10', 'D11']
    },
    index=[8, 9, 10, 11])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# concat soma o conjunto de dados
# concat(), CONCATENAR
# junta na horiental
pd.concat([df1, df2, df3])
# é possível concatenar pelas colunas, pelo axis(0 = índice, 1 = coluna)
# só utiliza axis = 1 quando os conjuntos compartilham a msm coluna
```

```python
esquerda = pd.DataFrame({
    'key' : ['K0', 'K1', 'K2', 'K3'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']})
```

```python
direita = pd.DataFrame({
    'key' : ['K0', 'K1', 'K2', 'K3'],
    'A' : ['C0', 'C1', 'C2', 'C3'],
    'B' : ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, on='key')
# é utilizado para juntar, msm tendo uma coluna igual nas datas
```

```python
direita = pd.DataFrame({
    'key1' : ['K0', 'K0', 'K1', 'K2'],
    'key2' : ['K0', 'K1', 'K0', 'K1'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']})
```

```python
esquerda = pd.DataFrame({
    'key1' : ['K0', 'K1', 'K1', 'K2'],
    'key2' : ['K0', 'K0', 'K0', 'K0'],
    'A' : ['A0', 'A1', 'C2', 'C3'],
    'B' : ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, on=['key1', 'key2'])
# procura ocorências dos pares que ocorrem de maneira simultânea na esquerda e na direita
```

```python
pd.merge(esquerda, direita, how= 'left', on=['key1', 'key2']) # é possível inserir um argumento how, que mostra de maneiras diferentes
# procura ocorências dos pares que ocorrem de maneira simultânea na esquerda e na direita
# é baseado nas colunas
```

```python
esquerda = pd.DataFrame({
    'A' : ['A0', 'A1', 'A2'],
    'B' : ['B0', 'B1', 'B2']},
    index= ['K0', 'K1', 'K2'])
```

```python
direita = pd.DataFrame({
    'C' : ['C0', 'C2', 'C3'],
    'D' : ['D1', 'D2', 'D3']},
    index= ['K0', 'K2', 'K3'])
```

```python
esquerda
```

```python
direita
```

```python
esquerda.join(direita)
# parece com merge, porém é voltado ao índice
```

### Operações com DataFrame:

```python
df.info()
# fala o tipo dos dados
```

```python
df.memory_usage()
# fala a quantidade de memória utilizada, em cada parte
```

```python
df['col2'].unique()
# retorna os que são únicos, tipo conjunto
```

```python
df['col2'].nunique()
# fala a quantidade de únicos
```

```python
df['col2'].value_counts()
# diz quantas vezes cada elemento se repete
```

```python
df
```

```python
def comp(x):
    return x ** 2 + 3
```

```python
data = {
    'A' : ['foo','foo', 'foo', 'bar', 'bar', 'bar'],
    'B' : ['one', 'one', 'two', 'two', 'one', 'one'],
    'C' : ['x', 'y', 'x', 'y', 'x', 'y'],
    'D' : [1, 3, 2, 5, 4, 1]
}
df = pd.DataFrame(data)
```

```python
df 
```

```python
dict_map = {
    'one' : 1,
    'two' : 2
}
# que legal, mapeia de acordo com o que aparece nas outras colunas
```

```python
df['E'] = df['B'].map(dict_map)
```

```python
df
```

```python
df.pivot_table(index= 'A', columns= 'B', values = 'D')
# é utilizado para criar uma tabela pivotada, se resume a uma nova forma de ordenar a linha de acordo com algo
```

### Serie temporal:

```python
# Serie temporal = é um conjunto de dados que em cada linha representa um momento diferente
# parece tranquilo de ser utilizado
```

```python
numeros_de_dias = 100
```

```python
datas = pd.date_range(start='1/1/2021', periods= numeros_de_dias)
# cria um array com diversas datas, partindo do start
```

```python
datas
```

```python
df = pd.DataFrame(range(numeros_de_dias), columns=['number'], index=datas)
```

```python
df
```

```python
df.index
# é um DateTime, python possui uma biblioteca DateTime
```

```python
df.index[0].day
```

```python
df.index[1].year
```

```python
df[df.index.day == 10]
```
