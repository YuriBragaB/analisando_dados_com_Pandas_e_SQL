### Iloc e filtros:

```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
df = pd.DataFrame(randn(5, 4), index = ['A', 'B', 'C', 'D', 'E'], columns='W X Y Z'.split( ))
```

```python
df['W']
# pega a coluna do DataFrame
# retorna uma Series
```

```python
df[['W', 'Y']]
# se botar uma lista pega mais de uma Series, retornando um DataFrame
```

```python
df['W']
# ainda retorna um DataFrame, msm sendo só um elemento, pois está contido em um array
```

```python
df.loc[['A', 'B']]
# pega pelo índice
```

```python
df.iloc[0, 2] 
# pega pela posição
# primeiro parâmetro = índice
# segundo parâmetro = coluna
```

```python
df.loc[['A', 'B'], 'W'] 
# tem um segundo parâmetro opcional, que adiciona mais um filtro por coluna
# retorna uma Series
# se botar um Array retorna um DataFrame
```

```python
df.iloc[1:, :]
# : seleciona tudo, é possível limitar os dois pontos, caso seja posto um número ao lado
# 1 tira o primeiro elemento
# -1 tira o último eleemnto
# muito parecido com o slicing dos arrays
```

```python
df.iloc[1:4, 1:3 ]
# retira linhas e colunas
```

```python
df > 0
# muito legal, é bem simples utilizar as condicionais
```

```python
df[df > 0]
# muito legal, poder passar na pesquisa uma condição
```

```python
df['Z'] > 0
# se for coluna retorna True os false
```

```python
df[df['W'] > 0]
# seleciona os dados da coluna W, retornando os positivo
```

```python
df[(df['W'] > 0) & (df['Y'] > 0)]
# [] = quando
# se não botar entre () retorna um valor booleano
```

### Hierarquia de índices e índices múltiplos

```python
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1, 2, 3, 1, 2, 3]

hier_index = list(zip(outside, inside))
# zip encaixa duas listas e retorna uma tupla
```

```python
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6, 2), index = hier_index, columns=['A', 'B'])
# faz duas partes de índice 
# é útil para fazer separações, tipo: homem e mulher
# é raro de se utilizar
```

```python
df
```

```python
df.loc['G1'].loc[1]
# para pegar o index numerico é necessário pegar a parte
```

```python
df.index.names = ['Grupo', 'Número']
```

```python
df
```

```python
df.xs(1, level = 'Número')
# pega o índice 1 das duas partes
```

### Manipulação de dados ausentes:

```python
df = pd.DataFrame({
    'A' : [1, 2, np.nan],
    'B' : [5, np.nan, np.nan],
    'C' : [1, 2, 3]
})
```

```python
df
```

```python
df.dropna()
# joga fora o índice 1 e 2, pois possui índices vazios
# drop NaN
```

```python
df.dropna(axis = 1)
# joga fora a coluna A e B
# pois possui NaN
```

```python
df.dropna(axis = 1, thresh = 2 )
# threh =  tolerância, nesse caso 1 NaN
```

```python
df.fillna(0)
# altera o valor dos NaN
# pode ser número ou string
```

```python
df['A'].mean()
# faz a média da Series
```

```python
df['A'].sum()
# soma a Series
```

```python
df['A'].fillna(value = df['A'].mean())
# preenche o NaN com a média dos valores da Series A
```

```python
df.ffill()
# preenche com o último número da coluna
```

### GroupBy:

```python
data = {
    'Classe': ['Júnior', 'Júnior', 'Pleno', 'Pleno', 'Sênior', 'Sênior'],
    'Nome' : ['Jorge', 'Carlos', 'Roberta', 'Patricia', 'Bruno', 'Vera'],
    'Venda' : [200, 120, 340, 124, 243, 350]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
group = df.groupby('Classe')
# guarde drupo em uma variável
# agrupa por classe
# tipo uma sanfona, agrupa dps expande
```

```python
group.sum(numeric_only=True)
# para aparecer apenas os números mude o numeric_only para True
```

```python
group.mean(numeric_only=True)
```

```python
df.groupby('Classe').min()
```

```python
df.groupby('Classe').max()
```

```python
# df2 = df
# se criar dessa forma dá erro, pois o python cria um ponteiro entre as duas, o correto é utilizar .copy()
df2 = df.copy()
```

```python
df2['Venda'] = [150, 432, 190, 230, 410,155]
```

```python
df
```

```python
df2
```

```python
df3 = pd.concat([df, df2])
# concatena os DataFrames
```

```python
df3
```

```python
df3.groupby(['Classe', 'Nome']).sum()
# da para ocupar mais de um elemento
```
