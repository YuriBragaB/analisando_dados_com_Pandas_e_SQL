### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_max = df3.groupby('ANO_MES').max(numeric_only=True)['PREÇO MÉDIO REVENDA']
df_min = df3.groupby('ANO_MES').min(numeric_only=True)['PREÇO MÉDIO REVENDA']

idx_max = df3.groupby('ANO_MES')['PREÇO MÉDIO REVENDA'].idxmax()
idx_min = df3.groupby('ANO_MES')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff['abs_diff'] = df_max - df_min
df_diff['percent_diff'] = (df_max - df_min) / df_min * 100
df_diff['max'] = df_max
df_diff['min'] = df_min
```

```python
df_diff['ESTADO_MAX'] = df3.loc[idx_max, :]['ESTADO'].values
df_diff['ESTADO_MIN'] = df3.loc[idx_min, :]['ESTADO'].values
```
